 ##Try fitting a line to a set of points

```
n =100
```

```
x = torch.ones(n,2)
x[:,0].uniform_(-1.,1)
x[:5]
```

```
a = tensor(3.,2);a
```

```
torch.rand(n)
```

```
y = x@a + torch.rand(n);y
```

```
plt.scatter(x[:,0],y)
```

```
def mse(y_pred,y):
  return ((y_pred-y)**2).mean()
```

```
a = tensor(-1.,1)
a
```

```
y_pred = x@a
y_pred
```

```
mse(y_pred, y)
```

```
plt.scatter(x[:,0],y)
plt.scatter(x[:,0],y_pred)
```

```
a = nn.Parameter(a)
a
```

```
def update():
  y_pred = x@a
  loss = mse(y,y_pred)
  if t % 10 == 0: print(loss)
  loss.backward()
  with torch.no_grad():
    a.sub_(lr*a.grad)
    a.grad.zero_()
  
```

```
lr = 1e-1
for t in range(100):
  update()
#a = a.detach().numpy()
y_pred = x@a.detach().numpy()
```

```
plt.scatter(x[:,0],y)
plt.scatter(x[:,0],y_pred);
```

```
from matplotlib import animation, rc
rc('animation', html='jshtml')
```

```
a = nn.Parameter(tensor(-1.,1))



fig = plt.figure()
plt.scatter(x[:,0],y)
line, = plt.plot(x[:,0], y_pred,c='red')

def animate(i):
  update()
  with torch.no_grad():
    y_pred = x@a
  line.set_ydata(y_pred)
  return line,

animation.FuncAnimation(fig, animate, np.arange(0,100), interval=20)
```
